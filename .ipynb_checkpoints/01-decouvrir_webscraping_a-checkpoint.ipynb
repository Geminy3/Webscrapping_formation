## Qu’est-ce que le webscraping ?

« récupération et organisation automatisées des données Web par leur extraction via un script ou un programme […] » source: [Wikipédia](https://fr.wikipedia.org/wiki/Web_scraping)

Autrement dit, c’est le passage d’une page HTML, sous un format interprétable et requêtable, à l’aide d’un script Python qui automatise le processus.


## Comment est construit le réseau internet ? 

Avant d'interagir avec le web il est préférable de connaître son fonctionnement. Voici quelques éléments du fonctionnement de l'architecture réseaux entre un ordinateur (client) et un site web (serveur): 

Un **client** (ordinateur, portable, imprimante, etc.) envoie une demande d’accès à une donnée au sein d’un **réseau local** (LAN). Ce réseau peut être organisé de différentes manières selon le nombre d’appareils connectés. Les données sont transmises via un câble **Ethernet** ou le **Wi-Fi**.
Si plusieurs appareils communiquent, les données transitent souvent par un **switch**, qui les achemine vers la bonne destination dans le LAN en utilisant les adresses **MAC** des appareils.

Pour sortir du LAN, les données passent par un routeur. Celui-ci utilise les **adresses IP** (l’"identifiant" des appareils) et le **protocole TCP** pour :

- Utilise les adresses IP (l’"identifiant" des appareils) et le protocole TCP (ou UDP) pour découper les données en paquets et les transmettre de manière fiable.

- Applique le NAT pour traduire les adresses IP privées du LAN en une adresse IP publique (fournie par le FAI).
  
- Peut inclure un pare-feu pour filtrer les connexions.

Le routeur se connecte ensuite au **FAI (Fournisseur d’Accès Internet)** via une technologie comme **PPPoE** ou la **fibre**. Le FAI achemine alors les paquets sur le **réseau étendu** (WAN), comme Internet.

Une fois arrivées à destination, les données sont échangées avec le serveur web via le protocole **HTTP** (ou HTTPS, sécurisé par TLS). Le serveur renvoie alors les données demandées (ex: page **HTML**, images).
Enfin, le navigateur du client affiche la page web reçue.

### Définitions

| Élément | Rôle |
| --- | --- 
| TCP | Garantit que les données arrivent sans erreur et dans l’ordre (comme un accusé de réception).
| IP | Achemine les données entre machines en utilisant des adresses (ex: 192.168.1.1).
| PPPoE | Connecte le routeur au FAI via une ligne Ethernet (comme une "poignée de main" pour accéder à Internet).
| HTTP | Affiche les pages web dans ton navigateur.
| FTP | Transfère des fichiers (upload/download) entre un client et un serveur.
| TLS | Chiffre les données pour sécuriser les échanges (ex: paiements en ligne, connexions distantes).
| API REST | Échange des données structurées (ex: une app mobile qui récupère des infos depuis un serveur).
| Routeur | Aiguille le trafic entre ton réseau local (LAN) et Internet (WAN). Traduit aussi les adresses privées en publiques (NAT).
| Switch | Relie les appareils de ton réseau local (ex: PC, imprimante) et transmet les données entre eux.

### Et maintenant une page Web ?

Une page web est composé de deux langages: 
- l'HTML, un langage à balise, pour construire le corps de la page
- CSS pour créer la forme de la page

Il est également possible d'intégrer un troisième langage, le Javascript. le Javascript permet de rendre des pages dynamiques. Il est aujourd'hui aussi incontourable que l'HTML et le CSS pour construire un site web.

Une page HTML est toujours construire de la même manière. Chaque élément est à l'intérieur de chevrons <> dans lequel il est possible de mettre différents attributs. 
En entête il est obligatoire de mettre ces éléments:
```html
<!DOCTYPE html>
<html lang="fr">
<head>
</head>
</html>
```
Il permettent d'indiquer qu'il s'agit d'un document html. Les balises sont la plupart d'un temps double, une balise ouvrante ```<html>``` et une balise fermante ```</html>```. À l'intérieur de ces balises, il est possible de mettre une chaîne de caractère comme ici ```<p>Une recette italienne classique avec des œufs, du parmesan et du guanciale.</p>```. 

```<p>``` est une balise qui indique qu'un paragraphe se trouve à l'intérieur. Cela permet de découper le texte. 



De nombreuses balises existent, aussi nous allons en présenter certaines:
  **Catégorie**         | **Balise**       | **Description**                                                                 | **Exemple**                                  | **Fermeture** |
 |-----------------------|------------------|---------------------------------------------------------------------------------|----------------------------------------------|---------------|
 | **Structure**         | `<!DOCTYPE html>` | Déclare le type de document (HTML5).                                          | `<!DOCTYPE html>`                            | Non           |
 |                       | `<html>`         | Racine du document HTML.                                                       | `<html>...</html>`                           | Oui           |
 |                       | `<head>`         | Contient les métadonnées (titres, styles, scripts).                           | `<head>...</head>`                           | Oui           |
 |                       | `<body>`         | Contient le contenu visible de la page.                                       | `<body>...</body>`                           | Oui           |
 |                       | `<title>`        | Titre de la page (affiché dans l'onglet du navigateur).                       | `<title>Ma Page</title>`                     | Oui           |
 |                       | `<meta>`         | Métadonnées (encodage, viewport, etc.).                                       | `<meta charset="UTF-8">`                     | Non           |
 | **Texte**            | `<h1>` à `<h6>`  | Titres de niveau 1 à 6.                                                        | `<h1>Titre principal</h1>`                  | Oui           |
 |                       | `<p>`            | Paragraphe.                                                                    | `<p>Un paragraphe.</p>`                      | Oui           |
 |                       | `<br>`           | Saut de ligne.                                                                 | `Texte<br>Nouvelle ligne`                    | Non           |        |
 |                       | `<strong>`       | Texte en **gras** (importance sémantique).                                     | `<strong>Important</strong>`                 | Oui           |
 |                       | `<b>`            | Texte en **gras** (style uniquement).                                         | `<b>Gras</b>`                                | Oui           |
 |                       | `<em>`           | Texte en *italique* (emphase sémantique).                                     | `<em>Emphase</em>`                            | Oui           |
 |                       | `<blockquote>`   | Citation longue.                                                               | `<blockquote>Citation</blockquote>`          | Oui           |
 |                       | `<q>`            | Citation courte (entre guillemets).                                           | `<q>Citation courte</q>`                     | Oui           |
 | **Listes**           | `<ul>`           | Liste non ordonnée.                                                           | `<ul><li>Élément</li></ul>`                  | Oui           |
 |                       | `<ol>`           | Liste ordonnée.                                                               | `<ol><li>Premier</li></ol>`                  | Oui           |
 |                       | `<li>`           | Élément de liste.                                                              | `<li>Élément</li>`                           | Oui           |
 | **Liens & Médias**   | `<a>`            | Lien hypertexte.                                                               | `<a href="url">Lien</a>`                     | Oui           |
 |                       | `<img>`          | Image.                                                                         | `<img src="image.jpg" alt="Description">`     | Non           |
 | **Tableaux**         | `<table>`        | Tableau.                                                                       | `<table>...</table>`                         | Oui           |
 | **Semantique**       | `<header>`       | En-tête de page ou section.                                                    | `<header>...</header>`                       | Oui           |
 |                       | `<footer>`       | Pied de page ou section.                                                       | `<footer>...</footer>`                       | Oui           |
 |                       | `<nav>`          | Section de navigation.                                                        | `<nav>...</nav>`                             | Oui           |
 | **Script & Style**   | `<script>`       | Code JavaScript.                                                               | `<script>alert("Hello");</script>`           | Oui           |
 |                       | `<style>`        | CSS intégré.                                                                   | `<style>body { color: red; }</style>`        | Oui           |
 |                       | `<link>`         | Lien vers une ressource externe (CSS, etc.).                                   | `<link rel="stylesheet" href="styles.css">`  | Non           |

En plus, des balises, quelques attributs, à l'intérieur de ces dernières, sont utiles à connaître:

  **Attribut**       | **Balises associées**       | **Rôle**                                                                                     | **Exemple**                                  | **Pourquoi c'est indispensable**                                                                 |
 |-------------------|-----------------------------|---------------------------------------------------------------------------------------------|---------------------------------------------|-----------------------------------------------------------------------------------------------|
 | **`id`**          | Toutes                      | Identifiant **unique** pour cibler un élément en CSS/JS.                                     | `<div id="header">`                         | Permet une sélection précise (ex: `#header { ... }` ou `document.getElementById()`).       |
 | **`class`**       | Toutes                      | Applique des **styles** ou des comportements à plusieurs éléments.                          | `<p class="text-center error">`              | Réutilisable (contrairement à `id`), base du styling moderne.                                  |
 | **`href`**        | `<a>`, `<link>`, `<area>`   | Définit la **destination** d'un lien ou d'une ressource externe.                            | `<a href="https://exemple.com">Lien</a>`    | Fondamental pour la navigation et l'intégration de CSS/ressources.                          |

## Mais avant de faire du webscraping, il est important de connaître les règles juridiques pour en faire

Le webscraping, est encadré par la loi française ainsi que des normes européennes. Avant de se lancer dans une tâche de webscraping, il est important de se renseigner sur les règles juridiques pour pouvoir le faire:

---

**Problématiques** :
- Droit de réutilisation des données.
- Absence d'information/consentement des personnes concernées.
- Risque de non-respect des droits (ex. : données personnelles).

---

### **1. Cadre Juridique en France**
##### **Textes applicables**
- **Code de la propriété intellectuelle** (Art. L. 342-3):
  - Autorise l'extraction/réutilisation d'une **partie non substantielle** (qualitative/quantitative) pour un accès licite.
  - Autorise l'extraction **à des fins privées** (respect des droits d'auteur).
  - Autorise l'extraction d'une **partie substantielle** **uniquement** pour l'**enseignement/recherche**, pour un public restreint (étudiants, chercheurs).

- **RGPD** :
  Applicable si les données sont **personnelles**.

- **Conditions Générales d'Utilisation (CGU)** :
  Certaines plateformes (ex. : LinkedIn) **interdisent** le scraping. Néanmoins, au regard de la hiérarchie des normes, la loi française prévaut sur les CGU des plateformes. 

---

### **2. Comment Scraper Légalement ?**
##### **Étapes clés**
1. **Consulter le DPO (Délégué à la Protection des Données)** :
   - Envoyer le projet pour analyse et inscription au registre.
   - Réaliser une **analyse d'impact** si nécessaire.
   - Définir :
     - Fondement juridique du projet.
     - Nécessité d'un **consentement** (obligatoire si données non publiques).
     - Modalités d'**information** des personnes (ex. : site du labo).

2. **Vérifier les CGU du site cible** :
   - Certaines plateformes interdisent explicitement le scraping.

3. **Respecter le RGPD** :
   - Informer les personnes concernées.
   - Obtenir leur **consentement** si les données ne sont pas publiques.

4. **Bonnes pratiques éthiques** :
   - Limiter les données collectées à l'**essentiel**.
   - Privilégier les **API** (si elles existent et qu'elles sont disponibles).
   - Respecter le fichier **`robots.txt`** (zones autorisées/interdites).
   - Éviter les requêtes massives (ne pas surcharger les serveurs).


*Exemple*: *Scraper les horaires de la SNCF est légal, mais scraper les profils LinkedIn nécessite une analyse RGPD.*

<br>

---
##### **3. Sanctions en Cas de Non-Respect**
   **Fondement**               | **Sanction**                                                                 |
 |-----------------------------|------------------------------------------------------------------------------|
 | **Droit pénal** (Art. 323-3) | Jusqu'à **150 000 € d'amende** et **5 ans d'emprisonnement**.               |
 | **Propriété intellectuelle** | Sanctions pour non-respect de l'Art. L. 342-3.                              |
 | **RGPD**                    | Amende jusqu'à **20M€** ou **4% du CA mondial** (CNIL).                     |

#### Les boucles

Les boucles permettent de répéter un certain nombre de fois une action.

Elles prennent deux formes : les boucles ***for*** et les boucles ***while***.

##### *While* : l'équivalent du "tant que"

Les boucles *while* permettent d'effectuer une action tant qu'une condition est remplie et donc égale à True.
 
On les utilise particulièrement lorsqu'on ne connait pas forcément à l'avance le nombre de fois qu'une action va être répétée avant que la condition soit remplie.

````python
i = 0
while (i < 10):
    print(i)
    i = i +1
````

Ci-dessus, nous connaissons déjà en soi le nombre de fois que la condition va s'éxecuter, mais observons un second exemple où cela n'est pas le cas...

````python
reponse = input("Entrez un mot : ")
while(reponse != "fini"):
    print(reponse)
    reponse = input("Entrez un mot : ")

print("Bravo ! Fin de programme.")
````

Attention, les boucles *while* nécessitent toujours de s'assurer qu'il existe une "sortie", un moment où la condition ne sera plus remplie et où la boucle pourra s'arrêter. Les boucles *while* peuvent en effet être en théorie infinie si leur condition est toujours égale à True...

```python
while True:
    print("Travail terminé!")
```

Ici la boucle se répéterait donc à l'infini...

##### ***For ... in*** : l'équivalent du "pour"
Les boucles *for* permettent aussi d'effectuer une action tant qu'une condition est remplie (égale à *True*).  
On les utilise plus souvent quand on connait déjà plus ou moins le nombre de répétition à réaliser.  
Le mot-clé *range* permet notamment de donner une plage de nombre.

````python
for i in range(0,15):
    print(i)
````

````python
for i in auteurs["Pénélope Bagieu"].keys():
    print(i)
````

*i* dans la boucle *for* correspond à un élément dans la variable *auteur* avec comme argument/filtre "Pénélope Bagieu" et plus précisent les clés (comme nous l'avons vu précédémment). Ensuite la boucle affiche chaque élément des clés correspondant à "Pénélope Bagieu" dans le dictionnaires des auteur·ices.

### Les fonctions

C’est un sous-ensemble d’instructions qui peut être appelé à des moments différents du script.
Il se caractérise par les deux mots-clés ***def*** et ***return*** : le premier, ***def***, signifiant *definition* permet de créer la fonction, le second de retourner (renvoyer) son résultat.  
***Return*** n’est pas obligatoire et ne sert que lorsqu'on veut retourner une résultat pour le stocker.

Le fonction prend souvent en compte des paramètres (c'est-à-dire des éléments obligatoires pour la faire fonctionner) mis entre parenthèses.


Il faut bien distinguer deux moments dans une fonction :
1. Le code qui créé la fonction mais ne l'exécute pas.
2. La commande qui appelle la fonction pour l'exécuter.

Exemple avec deux fonctions que vous connaissez déjà...
```python
print("Hello World")
len(disciplines)
```
En effet, ces deux commandes **appellent** les fonctions correspondantes (dont nous ne voyons pas le code exact qui est installé avec Python sur votre machine).  
Si nous prenons ***print*** en exemple, nous constatons donc le mot-clé ***print***, qui est le nom de la fonction, et des parenthèses qui encadrent des **paramètres** qui sont spécifiques à la fonction et décidés lors de la création de celle-ci. Ici, les créateurs de la fonction print ont décidé que la fonction afficherait la variable ou tout élément entrée comme paramètre.

Créons une fonction simple...


````python
def carre(n):
    new_nbr = n**2
    return new_nbr
````

La fonction ci-dessus s'appelle **carre** et prend un paramètre ***n*** : en s'exécutant, elle met n au carré dans une variable et la retourne.
Note à part, nous aurions pu simplifier en faisant simplement :
```python
def carre(n):
    return n**2
```
Les deux fonctions sont absolument identiques dans leur fonctionnement.

Maintenant, il ne reste plus qu'à l'**appeler** :

````python
carre(5)
````

````python
# Il est possible ensuite de conserver le résultat d'une fonction dans une variable
nb_a_garder = carre(3)
nb_a_garder
````

Pour notre premier test de moissonnage nous allons utiliser un site factice, fait pour être scraper: https://books.toscrape.com/



````python
# La première étape est de créer une variable est l'url du site à scraper afin de pouvoir l'appeler à volonté

url= 'https://books.toscrape.com/'
html=requests.get(url).text
````

````python
print(html) #html est une longue chaîne de caractères donc difficilement interrogeable.
````

````python
soup = BeautifulSoup(html, 'html.parser')   #soup est un objet complexe qui permet de naviguer dans l'arbre HTML
````

````python
print(soup)
````

````python
soup.title # on peut interroger par élément
````

````python
# il est également possible d'intérroger par type d'attributs
print(soup.ul['class'])
print(soup.a['href'])
````

````python
soup.find_all('a') #on peut rechercher tous les éléments à partir de leur nom
````

On voit que les genres littéraires sont présents l'intérieur de balise lien. On peut les récupérer directement sous forme de liste. 

````python
links = soup.find_all('ul',{"class":"nav-list"})[0].find_all('a') # pour cibler la recherche on commence par trouver tous les éléments à puce qui ont pour classe nav-list puis on cherche à l'intérieur les liens
categories = [link.get_text(strip=True) for link in links] # une fois qu'on a récupéré les liens, on peut faire une boucle pour récupérer le text à l'intérieur et obtenir une liste.
print(categories)
````
