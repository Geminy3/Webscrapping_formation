# Syntaxe de base Python


  **📌 Catégorie**       | **Syntaxe**               | **Résultat**                          | **Exemple**                     |
 |-----------------------|--------------------------|---------------------------------------|---------------------------------|
 | **🖨️ Affichage**      | `print(valeur)`          | Affiche une valeur dans la console.   | `print("Bonjour")` → `Bonjour`  |
 | **📦 Variables**       | `nom_variable = valeur`  | Assigne une valeur à une variable.   | `age = 25`                      |
 | **🔍 Types**          | `type(variable)`        | Retourne le type de la variable.      | `type(42)` → `<class 'int'>`    |
 |                       | `str(valeur)`           | Convertit en chaîne de caractères.    | `str(42)` → `"42"`              |
 |                       | `int(valeur)`           | Convertit en entier.                 | `int("42")` → `42`              |
 |                       | `float(valeur)`         | Convertit en nombre à virgule.        | `float("3.14")` → `3.14`        |
 |                       | `bool(valeur)`          | Convertit en booléen.                 | `bool(1)` → `True`              |
 | **⚙️ Opérateurs**     | `a + b`                | Addition.                             | `5 + 3` → `8`                   |
 |                       | `a - b`                | Soustraction.                         | `10 - 4` → `6`                  |
 |                       | `a * b`                | Multiplication.                       | `3 * 4` → `12`                  |
 |                       | `a / b`                | Division (résultat flottant).          | `10 / 3` → `3.333...`           |
 |                       | `a // b`               | Division entière.                    | `10 // 3` → `3`                 |
 |                       | `a % b`                | Modulo (reste de la division).        | `10 % 3` → `1`                  |
 |                       | `a ** b`               | Puissance.                            | `2 ** 3` → `8`                  |
 |                       | `a == b`               | Égalité.                              | `5 == 5` → `True`               |
 |                       | `a != b`               | Inégalité.                            | `5 != 3` → `True`               |
 |                       | `a < b`                | Inférieur à.                          | `3 < 5` → `True`                |
 |                       | `a <= b`               | Inférieur ou égal à.                  | `3 <= 3` → `True`               |
 |                       | `a > b`                | Supérieur à.                          | `5 > 3` → `True`                |
 |                       | `a >= b`               | Supérieur ou égal à.                  | `5 >= 5` → `True`               |
 |                       | `and`                  | ET logique.                           | `True and False` → `False`      |
 |                       | `or`                   | OU logique.                           | `True or False` → `True`        |
 |                       | `not`                  | NON logique.                          | `not True` → `False`            |
 | **💬 Commentaires**    | `# commentaire`        | Ajoute un commentaire (ignoré).        | `# Ceci est un commentaire`    |
 | **📋 Listes**          | `nom = []`             | Crée une liste vide.                  | `fruits = []`                   |
 |                       | `nom = [1, 2, 3]`      | Crée une liste avec des éléments.     | `nombres = [1, 2, 3]`           |
 |                       | `.append(x)`           | Ajoute `x` à la fin de la liste.      | `fruits.append("pomme")`        |
 |                       | `.insert(i, x)`        | Insère `x` à l'index `i`.             | `fruits.insert(0, "banane")`    |
 |                       | `del nom[i]`           | Supprime l'élément à l'index `i`.     | `del fruits[0]`                 |
 |                       | `.remove(x)`           | Supprime la première occurrence de `x`.| `fruits.remove("pomme")`        |
 |                       | `len(nom)`             | Retourne la longueur de la liste.     | `len([1, 2, 3])` → `3`          |
 | **📚 Dictionnaires**   | `nom = {}`             | Crée un dictionnaire vide.            | `personne = {}`                 |
 |                       | `nom = {"clé": "valeur"}` | Crée un dictionnaire.           | `personne = {"nom": "Alice"}`   |
 |                       | `nom["clé"] = valeur`  | Ajoute/modifie une clé/valeur.        | `personne["age"] = 30`          |
 |                       | `.keys()`              | Retourne les clés du dictionnaire.    | `personne.keys()` → `dict_keys(['nom', 'age'])` |
 |                       | `.values()`            | Retourne les valeurs du dictionnaire.| `personne.values()` → `dict_values(['Alice', 30])` |
 |                       | `.items()`             | Retourne les paires clé/valeur.       | `personne.items()` → `dict_items([('nom', 'Alice'), ('age', 30)])` |
 | **❓ Conditions**       | `if condition:`       | Exécute un bloc si la condition est vraie. | `if age >= 18: print("Majeur")` |
 |                       | `elif condition:`     | Sinon, si une autre condition est vraie. | `elif age >= 12: print("Adolescent")` |
 |                       | `else:`               | Sinon, exécute ce bloc.               | `else: print("Enfant")`          |
 | **🔁 Boucles**         | `for x in iterable:`   | Boucle sur chaque élément.            | `for fruit in ["pomme", "banane"]: print(fruit)` |
 |                       | `for i in range(a, b, step):` | Boucle avec un compteur. | `for i in range(3): print(i)` → `0, 1, 2` |
 |                       | `while condition:`    | Boucle tant que la condition est vraie. | `while x < 5: print(x); x += 1` |
 | **🎯 Fonctions**       | `def nom():`          | Définit une fonction.                | `def dire_bonjour(): print("Bonjour")` |
 |                       | `return valeur`       | Retourne une valeur.                 | `def carre(x): return x ** 2`   |
 |                       | `nom(arg1, arg2)`     | Appelle une fonction avec des arguments. | `carre(4)` → `16`          |
 | **🛡️ Erreurs**        | `try: ... except: ...` | Gère les exceptions (erreurs). | `try: print(10/0) except: print("Erreur !")` |
 | **🔎 Vérifications**   | `"clé" in nom`        | Vérifie si une clé existe dans un dictionnaire. | `"nom" in personne` → `True` |
 |                       | `"valeur" in nom.values()` | Vérifie si une valeur existe dans un dictionnaire. | `30 in personne.values()` → `True` |
 |                       | `x in liste`          | Vérifie si `x` est dans la liste.     | `"pomme" in fruits` → `True`     |
