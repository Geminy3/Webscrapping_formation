# Récapitulatif des principales balises HTML

  **Catégorie**         | **Balise**       | **Description**                                                                 | **Exemple**                                  | **Fermeture** |
 |-----------------------|------------------|---------------------------------------------------------------------------------|----------------------------------------------|---------------|
 | **Structure**         | `<!DOCTYPE html>` | Déclare le type de document (HTML5).                                          | `<!DOCTYPE html>`                            | Non           |
 |                       | `<html>`         | Racine du document HTML.                                                       | `<html>...</html>`                           | Oui           |
 |                       | `<head>`         | Contient les métadonnées (titres, styles, scripts).                           | `<head>...</head>`                           | Oui           |
 |                       | `<body>`         | Contient le contenu visible de la page.                                       | `<body>...</body>`                           | Oui           |
 |                       | `<title>`        | Titre de la page (affiché dans l'onglet du navigateur).                       | `<title>Ma Page</title>`                     | Oui           |
 |                       | `<meta>`         | Métadonnées (encodage, viewport, etc.).                                       | `<meta charset="UTF-8">`                     | Non           |
 | **Texte**            | `<h1>` à `<h6>`  | Titres de niveau 1 à 6.                                                        | `<h1>Titre principal</h1>`                  | Oui           |
 |                       | `<p>`            | Paragraphe.                                                                    | `<p>Un paragraphe.</p>`                      | Oui           |
 |                       | `<br>`           | Saut de ligne.                                                                 | `Texte<br>Nouvelle ligne`                    | Non           |
 |                       | `<hr>`           | Ligne horizontale (séparation thématique).                                    | `<hr>`                                        | Non           |
 |                       | `<strong>`       | Texte en **gras** (importance sémantique).                                     | `<strong>Important</strong>`                 | Oui           |
 |                       | `<b>`            | Texte en **gras** (style uniquement).                                         | `<b>Gras</b>`                                | Oui           |
 |                       | `<em>`           | Texte en *italique* (emphase sémantique).                                     | `<em>Emphase</em>`                            | Oui           |
 |                       | `<i>`            | Texte en *italique* (style uniquement).                                       | `<i>Italique</i>`                             | Oui           |
 |                       | `<mark>`         | Texte surligné (marqué).                                                       | `<mark>Surligné</mark>`                      | Oui           |
 |                       | `<small>`        | Texte en petit.                                                                | `<small>Petit texte</small>`                  | Oui           |
 |                       | `<sub>`          | Indice (exposant inférieur).                                                   | `H<sub>2</sub>O`                            | Oui           |
 |                       | `<sup>`          | Exposant.                                                                      | `x<sup>2</sup>`                              | Oui           |
 |                       | `<code>`         | Code source (monospace).                                                       | `<code>console.log()</code>`                 | Oui           |
 |                       | `<pre>`          | Texte préformaté (conserve les espaces et sauts de ligne).                     | `<pre>  Code\n  formaté</pre>`               | Oui           |
 |                       | `<blockquote>`   | Citation longue.                                                               | `<blockquote>Citation</blockquote>`          | Oui           |
 |                       | `<q>`            | Citation courte (entre guillemets).                                           | `<q>Citation courte</q>`                     | Oui           |
 | **Listes**           | `<ul>`           | Liste non ordonnée.                                                           | `<ul><li>Élément</li></ul>`                  | Oui           |
 |                       | `<ol>`           | Liste ordonnée.                                                               | `<ol><li>Premier</li></ol>`                  | Oui           |
 |                       | `<li>`           | Élément de liste.                                                              | `<li>Élément</li>`                           | Oui           |
 |                       | `<dl>`           | Liste de définitions.                                                          | `<dl><dt>Terme</dt><dd>Définition</dd></dl>` | Oui           |
 |                       | `<dt>`           | Terme dans une liste de définitions.                                           | `<dt>Terme</dt>`                             | Oui           |
 |                       | `<dd>`           | Définition dans une liste de définitions.                                     | `<dd>Définition</dd>`                        | Oui           |
 | **Liens & Médias**   | `<a>`            | Lien hypertexte.                                                               | `<a href="url">Lien</a>`                     | Oui           |
 |                       | `<img>`          | Image.                                                                         | `<img src="image.jpg" alt="Description">`     | Non           |
 |                       | `<audio>`        | Audio.                                                                         | `<audio src="audio.mp3" controls></audio>`    | Oui           |
 |                       | `<video>`        | Vidéo.                                                                         | `<video src="video.mp4" controls></video>`   | Oui           |
 |                       | `<source>`       | Source pour `<audio>` ou `<video>`.                                            | `<source src="video.mp4" type="video/mp4">`  | Non           |
 |                       | `<figure>`       | Conteneur pour une image, vidéo, etc. avec légende.                            | `<figure><img src="..."><figcaption>...</figcaption></figure>` | Oui |
 |                       | `<figcaption>`   | Légende pour `<figure>`.                                                       | `<figcaption>Légende</figcaption>`           | Oui           |
 | **Tableaux**         | `<table>`        | Tableau.                                                                       | `<table>...</table>`                         | Oui           |
 |                       | `<tr>`           | Ligne de tableau.                                                              | `<tr><td>Cellule</td></tr>`                  | Oui           |
 |                       | `<th>`           | En-tête de tableau.                                                            | `<th>En-tête</th>`                           | Oui           |
 |                       | `<td>`           | Cellule de tableau.                                                            | `<td>Donnée</td>`                            | Oui           |
 |                       | `<thead>`        | Groupe d'en-têtes de tableau.                                                  | `<thead><tr><th>...</th></tr></thead>`       | Oui           |
 |                       | `<tbody>`        | Corps du tableau.                                                              | `<tbody>...</tbody>`                         | Oui           |
 |                       | `<tfoot>`        | Pied de tableau.                                                               | `<tfoot>...</tfoot>`                         | Oui           |
 | **Formulaires**      | `<form>`         | Formulaire.                                                                    | `<form action="/submit" method="post">...</form>` | Oui       |
 |                       | `<input>`        | Champ de saisie (texte, mot de passe, etc.).                                   | `<input type="text" name="nom">`              | Non           |
 |                       | `<label>`        | Étiquette pour un champ de formulaire.                                         | `<label for="nom">Nom :</label>`             | Oui           |
 |                       | `<textarea>`     | Zone de texte multiligne.                                                      | `<textarea name="message"></textarea>`       | Oui           |
 |                       | `<button>`       | Bouton.                                                                         | `<button type="submit">Envoyer</button>`     | Oui           |
 |                       | `<select>`       | Liste déroulante.                                                              | `<select><option>Option</option></select>`   | Oui           |
 |                       | `<option>`       | Option dans une liste déroulante.                                              | `<option value="val">Option</option>`        | Oui           |
 |                       | `<optgroup>`     | Groupe d'options dans une liste déroulante.                                    | `<optgroup label="Groupe">...</optgroup>`    | Oui           |
 |                       | `<fieldset>`     | Groupe de champs de formulaire.                                                | `<fieldset><legend>Titre</legend>...</fieldset>` | Oui       |
 |                       | `<legend>`       | Titre pour `<fieldset>`.                                                       | `<legend>Informations</legend>`              | Oui           |
 | **Semantique**       | `<header>`       | En-tête de page ou section.                                                    | `<header>...</header>`                       | Oui           |
 |                       | `<footer>`       | Pied de page ou section.                                                       | `<footer>...</footer>`                       | Oui           |
 |                       | `<nav>`          | Section de navigation.                                                        | `<nav>...</nav>`                             | Oui           |
 |                       | `<main>`         | Contenu principal de la page.                                                 | `<main>...</main>`                           | Oui           |
 |                       | `<article>`      | Contenu autonome (article, commentaire, etc.).                                | `<article>...</article>`                     | Oui           |
 |                       | `<section>`      | Section thématique.                                                            | `<section>...</section>`                     | Oui           |
 |                       | `<aside>`        | Contenu secondaire (barre latérale, etc.).                                     | `<aside>...</aside>`                         | Oui           |
 |                       | `<div>`          | Conteneur générique (sans sémantique).                                         | `<div>...</div>`                             | Oui           |
 |                       | `<span>`         | Conteneur générique en ligne.                                                  | `<span>...</span>`                           | Oui           |
 | **Script & Style**   | `<script>`       | Code JavaScript.                                                               | `<script>alert("Hello");</script>`           | Oui           |
 |                       | `<style>`        | CSS intégré.                                                                   | `<style>body { color: red; }</style>`        | Oui           |
 |                       | `<link>`         | Lien vers une ressource externe (CSS, etc.).                                   | `<link rel="stylesheet" href="styles.css">`  | Non           |
 | **Autres**           | `<!-- -->`      | Commentaire (non affiché).                                                     | `<!-- Commentaire -->`                        | Non           |
 |                       | `<iframe>`       | Intègre une page web externe.                                                  | `<iframe src="url"></iframe>`                | Oui           |
 |                       | `<details>`      | Section dépliable.                                                             | `<details><summary>Titre</summary>...</details>` | Oui       |
 |                       | `<summary>`      | Titre pour `<details>`.                                                        | `<summary>Cliquez pour afficher</summary>`   | Oui           |
 |                       | `<progress>`     | Barre de progression.                                                          | `<progress value="50" max="100"></progress>` | Oui           |
 |                       | `<meter>`        | Jauge (plage de valeurs).                                                      | `<meter value="0.6">60%</meter>`            | Oui           |

# Principaux attributs des balises HTML

  **Catégorie**               | **Attribut**               | **Description**                                                                                     | **Balises associées**                          | **Valeurs/Exemples**                                                                                     | **HTML5** |
 |-----------------------------|----------------------------|-----------------------------------------------------------------------------------------------------|-----------------------------------------------|----------------------------------------------------------------------------------------------------------|-----------|
 | **⚡ Attributs globaux**     | `id`                       | Identifiant **unique** pour un élément (utilisé en CSS/JS).                                         | Toutes                                         | `id="header"`                                                                                         | Oui       |
 |                             | `class`                    | Classe(s) pour le style (CSS) ou la sélection (JS).                                               | Toutes                                         | `class="btn primary"` (plusieurs classes séparées par des espaces)                                   | Oui       |
 |                             | `style`                    | Styles CSS **en ligne**.                                                                           | Toutes                                         | `style="color: red; font-size: 16px;"`                                                                | Oui       |
 |                             | `title`                    | Info-bulle (tooltip) au survol.                                                                     | Toutes                                         | `title="Ceci est un bouton"`                                                                          | Oui       |
 |                             | `lang`                     | Langue du contenu.                                                                                 | Toutes (surtout `<html>`, `<p>`, `<span>`)      | `lang="fr"`, `lang="en-US"`                                                                          | Oui       |
 |                             | `data-*`                  | Attributs personnalisés (pour JavaScript).                                                         | Toutes                                         | `data-user-id="123"`, `data-role="admin"` (accessibles via `element.dataset.userId`)               | Oui       |
 |                             | `tabindex`                 | Ordre de tabulation (accessibilité).                                                               | Toutes                                         | `tabindex="1"` (focusable), `tabindex="-1"` (non focusable mais accessible via JS)                     | Oui       |
 |                             | `hidden`                   | Masque l'élément (équivalent à `display: none;` en CSS).                                           | Toutes                                         | `hidden` (pas de valeur)                                                                             | Oui       |
 |                             | `dir`                      | Direction du texte.                                                                               | Toutes                                         | `dir="ltr"` (gauche à droite), `dir="rtl"` (droite à gauche), `dir="auto"`                           | Oui       |
 |                             | `accesskey`                | Raccourci clavier pour accéder à l'élément.                                                        | Toutes (sauf `<base>`, `<bdo>`, `<br>`, etc.) | `accesskey="s"` (Alt+S sur Windows)                                                                   | Oui       |
 |                             | `draggable`                | Autorise le glisser-déposer.                                                                       | Toutes                                         | `draggable="true"` ou `false`                                                                         | Oui       |
 |                             | `spellcheck`               | Active la vérification orthographique.                                                           | `<input>`, `<textarea>`, éléments éditables   | `spellcheck="true"` (défaut) ou `false`                                                               | Oui       |
 | **🔗 Liens & Navigation**   | `href`                      | URL de destination pour un lien.                                                                  | `<a>`, `<link>`, `<area>`                     | `href="https://example.com"`, `href="#section"`, `href="mailto:contact@example.com"`               | Oui       |
 |                             | `target`                    | Où ouvrir le lien.                                                                                 | `<a>`, `<form>`, `<area>`                     | `_self` (défaut), `_blank` (nouvel onglet), `_parent`, `_top`, `nom-de-fenêtre`                     | Oui       |
 |                             | `rel`                       | Relation entre la page et la ressource liée.                                                     | `<a>`, `<link>`, `<area>`                     | `rel="nofollow"` (SEO), `rel="noopener noreferrer"` (sécurité pour `_blank`), `rel="stylesheet"` (CSS) | Oui       |
 |                             | `download`                  | Force le téléchargement au clic.                                                                   | `<a>`, `<area>`                                | `download` ou `download="nom-fichier.pdf"`                                                          | Oui       |
 |                             | `hreflang`                  | Langue de la ressource liée.                                                                       | `<a>`, `<link>`, `<area>`                     | `hreflang="en"`                                                                                      | Oui       |
 | **🖼️ Images & Médias**      | `src`                       | Source (URL) de l'image/média.                                                                     | `<img>`, `<audio>`, `<video>`, `<source>`, `<iframe>`, `<script>` | `src="image.jpg"`                                                                                   | Oui       |
 |                             | `alt`                       | Texte alternatif (accessibilité).                                                                 | `<img>`, `<area>`                              | `alt="Description de l'image"` (obligatoire pour l'accessibilité)                                  | Oui       |
 |                             | `width` / `height`           | Dimensions de l'image/média (en pixels).                                                          | `<img>`, `<video>`, `<canvas>`, `<svg>`       | `width="200"`, `height="auto"`                                                                       | Oui       |
 |                             | `loading`                   | Contrôle le chargement différé (lazy loading).                                                    | `<img>`, `<iframe>`                           | `loading="lazy"` (chargement au scroll), `eager` (défaut)                                             | Oui       |
 |                             | `decoding`                  | Mode de décodage de l'image.                                                                       | `<img>`                                        | `decoding="async"` (défaut), `sync`, `auto`                                                          | Oui       |
 |                             | `srcset`                    | Liste de sources pour les images responsives.                                                     | `<img>`, `<source>` (dans `<picture>`)          | `srcset="image-480w.jpg 480w, image-800w.jpg 800w"`                                                  | Oui       |
 |                             | `sizes`                     | Taille de l'image en fonction de la largeur de la viewport.                                       | `<img>`, `<source>`                            | `sizes="(max-width: 600px) 480px, 800px"`                                                           | Oui       |
 |                             | `crossorigin`                | Contrôle les requêtes CORS pour les images.                                                        | `<img>`, `<script>`, `<link>`, `<video>`      | `crossorigin="anonymous"`, `use-credentials`                                                        | Oui       |
 |                             | `ismap` / `usemap`           | Images cliquables (side server/client).                                                          | `<img>`                                        | `ismap` (serveur), `usemap="#map-name"` (client, avec `<map>`)                                       | Oui       |
 |                             | `controls`                  | Affiche les contrôles pour les médias.                                                            | `<audio>`, `<video>`                          | `controls` (pas de valeur)                                                                          | Oui       |
 |                             | `autoplay`                   | Lecture automatique.                                                                               | `<audio>`, `<video>`                          | `autoplay` (souvent bloqué par les navigateurs sans `muted`)                                         | Oui       |
 |                             | `loop`                       | Lecture en boucle.                                                                                 | `<audio>`, `<video>`                          | `loop`                                                                                              | Oui       |
 |                             | `muted`                      | Désactive le son.                                                                                 | `<audio>`, `<video>`                          | `muted`                                                                                              | Oui       |
 |                             | `poster`                     | Image de prévisualisation pour les vidéos.                                                        | `<video>`                                     | `poster="poster.jpg"`                                                                               | Oui       |
 |                             | `preload`                    | Préchargement du média.                                                                             | `<audio>`, `<video>`                          | `preload="none"` (désactivé), `metadata` (métadonnées uniquement), `auto` (défaut)                 | Oui       |
 | **📝 Formulaires**          | `action`                     | URL de soumission du formulaire.                                                                   | `<form>`                                      | `action="/submit"`                                                                                   | Oui       |
 |                             | `method`                     | Méthode HTTP pour la soumission.                                                                  | `<form>`                                      | `method="get"` (défaut), `post`, `dialog` (HTML5)                                                    | Oui       |
 |                             | `enctype`                    | Type de codage des données du formulaire.                                                         | `<form>`                                      | `enctype="application/x-www-form-urlencoded"` (défaut), `multipart/form-data` (fichiers), `text/plain` | Oui       |
 |                             | `name`                       | Nom du champ (utilisé pour la soumission).                                                        | `<input>`, `<textarea>`, `<select>`, `<button>`, `<form>`, `<fieldset>`, `<output>` | `name="username"`                                                                               | Oui       |
 |                             | `value`                      | Valeur par défaut ou soumise.                                                                      | `<input>`, `<option>`, `<button>`, `<textarea>`, `<select>` | `value="Soumettre"` (pour `<button>`), `value="option1"` (pour `<option>`)                        | Oui       |
 |                             | `type`                       | Type de champ de formulaire.                                                                       | `<input>`, `<button>`                         | `type="text"`, `password`, `email`, `number`, `date`, `checkbox`, `radio`, `file`, `submit`, `hidden` | Oui       |
 |                             | `placeholder`                 | Texte d'exemple (disparaît à la saisie).                                                          | `<input>`, `<textarea>`                      | `placeholder="Entrez votre nom"`                                                                   | Oui       |
 |                             | `required`                   | Champ obligatoire.                                                                               | `<input>`, `<textarea>`, `<select>`           | `required`                                                                                          | Oui       |
 |                             | `readonly`                   | Champ en lecture seule (modifiable via JS).                                                       | `<input>`, `<textarea>`                      | `readonly`                                                                                          | Oui       |
 |                             | `disabled`                   | Désactive le champ (non modifiable, non soumis).                                                  | `<input>`, `<textarea>`, `<select>`, `<button>`, `<fieldset>`, `<optgroup>`, `<option>` | `disabled` | Oui       |
 |                             | `autofocus`                   | Focus automatique au chargement de la page.                                                       | `<input>`, `<textarea>`, `<select>`, `<button>` | `autofocus`                                                                                     | Oui       |
 |                             | `autocomplete`                | Active/désactive l'autocomplétion.                                                                | `<input>`, `<textarea>`, `<select>`, `<form>` | `autocomplete="on"` (défaut), `off`, `name`, `email`, `address-level1` (pays), etc.               | Oui       |
 |                             | `min` / `max`                 | Valeurs minimale/maximale (pour `type="number"`, `date`, etc.).                                   | `<input>`                                     | `min="0"`, `max="100"`                                                                              | Oui       |
 |                             | `step`                       | Incrément pour les champs numériques.                                                             | `<input type="number" step="0.1">`          | `step="1"` (défaut), `step="0.01"`                                                                  | Oui       |
 |                             | `pattern`                    | Expression régulière pour validation.                                                            | `<input type="text">`                         | `pattern="[A-Za-z]+"` (lettres uniquement)                                                        | Oui       |
 |                             | `size`                       | Taille visible du champ (en caractères).                                                          | `<input>`, `<select>`                         | `size="20"`                                                                                         | Oui       |
 |                             | `maxlength`                   | Nombre maximal de caractères.                                                                     | `<input>`, `<textarea>`                      | `maxlength="50"`                                                                                   | Oui       |
 |                             | `minlength`                   | Nombre minimal de caractères.                                                                     | `<input>`, `<textarea>`                      | `minlength="5"`                                                                                     | Oui       |
 |                             | `multiple`                   | Autorise plusieurs valeurs.                                                                     | `<input type="file"|email">`, `<select>`      | `multiple`                                                                                         | Oui       |
 |                             | `checked`                    | Case cochée par défaut.                                                                           | `<input type="checkbox"|radio">`               | `checked`                                                                                          | Oui       |
 |                             | `selected`                    | Option sélectionnée par défaut.                                                                  | `<option>`                                    | `selected`                                                                                          | Oui       |
 |                             | `cols` / `rows`               | Dimensions d'un `<textarea>` (colonnes/lignes).                                                   | `<textarea>`                                 | `cols="50"`, `rows="5"`                                                                             | Oui       |
 |                             | `wrap`                       | Comportement du retour à la ligne dans `<textarea>`.                                              | `<textarea>`                                 | `wrap="soft"` (défaut, retour à la ligne dans l'UI mais pas dans la valeur), `hard` (inclus dans la valeur) | Oui       |
 |                             | `for`                        | Associe un `<label>` à un champ de formulaire.                                                    | `<label>`                                    | `for="username"` (doit correspondre à l'`id` du champ)                                              | Oui       |
 | **📊 Tableaux**             | `colspan`                     | Fusionne des cellules horizontalement.                                                           | `<td>`, `<th>`                               | `colspan="2"`                                                                                      | Oui       |
 |                             | `rowspan`                     | Fusionne des cellules verticalement.                                                             | `<td>`, `<th>`                               | `rowspan="3"`                                                                                      | Oui       |
 |                             | `scope`                       | Définit la portée d'un en-tête de tableau (accessibilité).                                       | `<th>`                                       | `scope="col"` (colonne), `row` (ligne), `colgroup`, `rowgroup`                                      | Oui       |
 | **🎨 Styles & Scripts**      | `rel`                         | Relation entre le document et la ressource.                                                     | `<link>`, `<a>`, `<area>`                     | `rel="stylesheet"` (CSS), `rel="icon"` (favicon), `rel="preload"` (préchargement)                | Oui       |
 |                             | `type`                        | Type de la ressource.                                                                             | `<script>`, `<style>`, `<link>`, `<object>`, `<source>` | `type="text/javascript"` (défaut pour `<script>`), `type="text/css"` (pour `<style>`)           | Oui       |
 |                             | `async` / `defer`              | Contrôle le chargement des scripts.                                                               | `<script>`                                   | `async` (exécution asynchrone), `defer` (exécution après le DOM)                                  | Oui       |
 |                             | `integrity`                   | Vérification de l'intégrité des ressources (SRI).                                               | `<script>`, `<link>`                          | `integrity="sha384-..."` + `crossorigin="anonymous"`                                               | Oui       |
 |                             | `nonce`                       | Autorise l'exécution d'un script (CSP).                                                          | `<script>`, `<style>`                        | `nonce="EDNnf03nceIOfn39fn3e9h3sdfa"`                                                             | Oui       |
 | **📱 Responsif & Médias**   | `media`                        | Requête média pour les styles ou scripts.                                                        | `<link>`, `<style>`, `<source>` (dans `<picture>`) | `media="(max-width: 600px)"`                                                                       | Oui       |
 |                             | `sizes`                        | Taille de l'image en fonction de la viewport (pour `<source>` dans `<picture>`).                  | `<source>` (dans `<picture>`)                 | `sizes="(max-width: 600px) 480px, 800px"`                                                           | Oui       |
 | **🔌 Iframes & Intégrations** | `sandbox`                     | Restrictions de sécurité pour les iframes.                                                       | `<iframe>`                                   | `sandbox=""` (toutes les restrictions), `sandbox="allow-scripts allow-forms"`                  | Oui       |
 |                             | `allow`                       | Autorise des fonctionnalités spécifiques dans une iframe.                                      | `<iframe>`                                   | `allow="geolocation; microphone"` (pour les APIs)                                                 | Oui       |
 |                             | `allowfullscreen`             | Autorise le mode plein écran.                                                                     | `<iframe>`                                   | `allowfullscreen`                                                                                 | Oui       |
 |                             | `frameborder`                 | **Dépécié** : Bordure de l'iframe (utilisez CSS `border` à la place).                             | `<iframe>`                                   | `frameborder="0"` (plus de bordure)                                                              | Non        |
 | **📌 Métadonnées**          | `charset`                      | Encodage des caractères.                                                                          | `<meta>`                                     | `charset="UTF-8"`                                                                                  | Oui       |
 |                             | `name` / `content`             | Nom et contenu d'une méta-donnée.                                                                | `<meta>`                                     | `<meta name="description" content="Ma page">`                                                     | Oui       |
 |                             | `http-equiv`                  | Équivalent HTTP pour les méta-données.                                                           | `<meta>`                                     | `<meta http-equiv="X-UA-Compatible" content="IE=edge">`                                         | Oui       |
 |                             | `viewport`                     | Contrôle l'affichage sur mobile.                                                                   | `<meta>`                                     | `<meta name="viewport" content="width=device-width, initial-scale=1.0">`                          | Oui       |
 | **🎯 Événements**           | `on*` (ex: `onclick`)         | Attributs d'événements (en ligne). **À éviter** (préférez `addEventListener`).                   | Toutes (sauf quelques exceptions)           | `onclick="alert('Cliqué !')"`                                                                     | Oui       |
 |                             | Exemples :                    | `onload`, `onerror`, `onfocus`, `onblur`, `onchange`, `onsubmit`, `onkeydown`, `onmouseover`, etc. | Toutes                                         | `onload="init()"`                                                                                 | Oui       |

# Principales propriétés et selecteurs CSS

  **Catégorie**               | **Propriété/Sélecteur**       | **Description**                                                                                     | **Valeurs/Exemples**                                                                                     | **Héritage** |
